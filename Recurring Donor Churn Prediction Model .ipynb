# Recurring Donor Churn Prediction

Revenue from recurring donors (sustainers that give monthly recurring gifts) are a key way the (non-profit)organization gets revenue from its members to power the work to be done. Hence, it is important to be able to have an idea of who is most likely to churn, and possible predictors of churn so that the relevant team can experiment with different intervention strategies to minimize the number of recurring donors that churn. 

# Importing and Initial Pre-Processing of Data

```python
#Importing Packages
import civis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
#from sklearn.datasets import make_classification

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report, accuracy_score,confusion_matrix, precision_score, recall_score

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

import shap  

client = civis.APIClient()
```

```python
#Importing dataset from RedShift
predictor = civis.io.read_civis('coc_reporting.sustainer_retention_prediction', 'TMC', use_pandas=True)

#Converting to date-time columns for particular columns
predictor['my_date_gift_started'] = pd.to_datetime(predictor['my_date_gift_started'], errors='coerce')
predictor['my_last_gift'] = pd.to_datetime(predictor['my_last_gift'], errors='coerce')
predictor['date_gift_started'] = pd.to_datetime(predictor['date_gift_started'], errors='coerce')

#Preliminary feature engineering 
predictor['gift_started_year'] = predictor['date_gift_started'].dt.year
predictor['gift_started_month'] = predictor['date_gift_started'].dt.month
predictor['gift_started_day'] = predictor['date_gift_started'].dt.day
predictor['opps_vs_action_ratio'] = predictor['total_mailings_received']*1.0/predictor['actions_before_last_gift']

#Looking at the first 5 columns
predictor.head()
```

# Exploratory Data Analysis 

```python
#Checking the balance of the classes

g = sns.countplot(predictor['lapsed'])
g.set_xticklabels(['Not lapsed','Lapsed'])
plt.show()
```

```python
names = ['actions_before_last_gift','num_actions_month_before_lg', 'total_mailings_received','num_mailings_received_month_before_lg']
ncols = len(names)
fig, axes = plt.subplots(1,ncols)

for name, ax in zip(names, axes.flatten()):
    sns.boxplot(x='lapsed', y= name, data=predictor, orient='v', ax=ax)

plt.subplots_adjust(right=3)
plt.show()
```

Since this is an extension of the recurring donor cohort grids analysis, I decided to focus on exploring some features that I believed would be relevant to predict churn, and see how they differed between lapsed and active recurring donors. 

The features I chose were:
1. Total actions the recurring donor took before their last recurring gift (in the series)
2. Number of actions the recurring donor took in the month before their last recurring gift 
3. Total mailings the recurring donor received before their last recurring gift
4. Number of mailings the recurring donor received in the month before their last recurring gift

Though the box plots above show different distributions across the four chosen features, it is interesting to note that the median for lapsed and active recurring donors are roughly the same across the chosen features.

# Feature Engineering

```python
#Dropping columns that won't be relevant for the models
dataset = predictor.drop(['last_gift', 'my_last_gift','user_id', 'order_id'
                          , 'date_gift_started', 'my_date_gift_started'], axis=1)
```

```python
#Checking to see if there are any highly correlated features so they can be dropped
#If any set of features have a correlation >= +0.5/-0.5, then one of the features would be dropped

dataset.corr()
```

I chose to drop the total_mailings_received feature because there is already another feature in the dataset that is related to mailings already (number of mailings received in the month before their last recurring gift)

```python
dataset = dataset.drop(['total_mailings_received'], axis=1)

#One-hot encoding date-time variables
dataset = pd.get_dummies(dataset, columns=['gift_started_year'], prefix='year') 
dataset = pd.get_dummies(dataset, columns=['gift_started_month'], prefix='month')
dataset = pd.get_dummies(dataset, columns=['gift_started_day'], prefix='day')
```

# Modelling and Evaluation

```python
#Dividing dataset into training and test set

X = dataset.drop(['lapsed'], axis=1)
y = dataset['lapsed']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#Scaling features so they would not skew the model's predictions; fitting to train set
ablg_trans = MinMaxScaler(feature_range = (0,1)).fit(np.array(X_train['actions_before_last_gift']).reshape(-1, 1))
namblg_trans = MinMaxScaler(feature_range = (0,1)).fit(np.array(X_train['num_actions_month_before_lg']).reshape(-1, 1))
nmrmblg_trans = MinMaxScaler(feature_range = (0,1)).fit(np.array(X_train['num_mailings_received_month_before_lg']).reshape(-1, 1))
ovar_trans = MinMaxScaler(feature_range = (0,1)).fit(np.array(X_train['opps_vs_action_ratio']).reshape(-1, 1))
```

```python
#Transforming training set
X_train['actions_before_last_gift'] = ablg_trans.transform(np.array(X_train['actions_before_last_gift']).reshape(-1, 1))
X_train['num_actions_month_before_lg'] = namblg_trans.transform(np.array(X_train['num_actions_month_before_lg']).reshape(-1, 1))
X_train['num_mailings_received_month_before_lg'] = nmrmblg_trans.transform(np.array(X_train['num_mailings_received_month_before_lg']).reshape(-1, 1))
X_train['opps_vs_action_ratio'] = ovar_trans .transform(np.array(X_train['opps_vs_action_ratio']).reshape(-1, 1))
```

```python
#Scaling and transforming test set
X_test['actions_before_last_gift'] = ablg_trans.transform(np.array(X_test['actions_before_last_gift']).reshape(-1, 1))
X_test['num_actions_month_before_lg'] = namblg_trans.transform(np.array(X_test['num_actions_month_before_lg']).reshape(-1, 1))
X_test['num_mailings_received_month_before_lg'] = nmrmblg_trans.transform(np.array(X_test['num_mailings_received_month_before_lg']).reshape(-1, 1))
X_test['opps_vs_action_ratio'] = ovar_trans .transform(np.array(X_test['opps_vs_action_ratio']).reshape(-1, 1))
```

```python
#Running the different classification models

models = {}

# Logistic Regression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
models['Support Vector Machines'] = LinearSVC()

# Decision Trees
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
models['K-Nearest Neighbor'] = KNeighborsClassifier()

#XGBoost
models['XGBoost'] = XGBClassifier()
```

```python
#Compiling evaluation metrics

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    # Fit the classifier model
    models[key].fit(X_train, y_train)
    
    # Prediction 
    predictions = models[key].predict(X_test)
    
    # Calculate Accuracy, Precision and Recall Metrics
    accuracy[key] = accuracy_score(y_test,predictions)
    precision[key] = precision_score(y_test,predictions)
    recall[key] = recall_score(y_test,predictions)
    
df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()

df_model
```

The most important metric here to look out for is **recall**, because we would want to minimize **false negatives** as much as possible (Type 2 error rate). False negative are the people that the model predicts would not lapse, but actually lapse, and this is not helpful because those are the people we are trying to retain.

Hence, **Support Vector Machine (SVM)** is the best performing model. 

## Using SMOTE for imbalanced sampling 

As we saw earlier, the lapsed and active recurring donor classes are imbalanced - active recurring donors are 30%. If I wrongly chose accuracy as the primary evaluation metric, that might have affected the prediction and the choice of the best performing model.

Since we are focusing on improving recall, SMOTE is appropriate to solve the class imbalance. SMOTE helps if we are focusing on improving the recall of the models, by overssampling the minority class in the dataset, before making a prediction. 

```python
#Running SMOTE on training set

oversample = SMOTE()
X_train_smote, y_train_smote = oversample.fit_resample(X_train, y_train)
```

```python
#Running the different classification models

models = {}

# Logistic Regression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
models['Support Vector Machines'] = LinearSVC()

# Decision Trees
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
models['K-Nearest Neighbor'] = KNeighborsClassifier()

# XGBoost
models['XGBoost'] = XGBClassifier()
```

```python
#Compiling evaluation metrics

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    # Fit the classifier model
    models[key].fit(X_train_smote, y_train_smote)
    
    # Prediction 
    predictions = models[key].predict(X_test)
    
    # Calculate Accuracy, Precision and Recall Metrics
    accuracy[key] = accuracy_score(y_test,predictions)
    precision[key] = precision_score(y_test,predictions)
    recall[key] = recall_score(y_test,predictions)
    
df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()

df_model
```

Looking at recall, SMOTE reduced the performance of the models, which was the opposite of what we would expect, and it is likely because the parameters were not optimised. Here, **random forest** is the best perfoming model based on recall.

However, SMOTE improved the precision of the models.

# Feature Importance in Predictions

## Linear Support Vector Machine

**Linear SVM** can allow us see what features were the most important for the model's prediction, which would be helpful in giving some direction on some possible things to implement to improve recurring donor retention. I chose this model because it was the best performing model before implementing SMOTE.

```python
classifier = LinearSVC()
classifier.fit(X_train, y_train) 
predictions = classifier.predict(X_test)

feat_importances = pd.Series(abs(classifier.coef_[0]), index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
```

The graph above shows that the **number of actions the recurring donor took in the month before their last gift** was the most influential feature in predicting whether a recurring donor would churn or not. This makes logical business sense to be one of the strongest predictors of whether a recurring donor lapses or not, because we would expect more active members who take actions frequently with the org to be less likely to lapse.

Though we can't predict or influence people's actions, we can influence the number of mailings people get and the type of content they get. We can run experiments with varying email content and targeting as new recurring donors give gifts to see how we can improve recurring donor retention using improved content that is more tailored to what interests the recurring donor. 

## Random Forests

**Random forests** can also allow us see what features were the most important for the model's prediction, which would be helpful in giving some direction on some possible things to implement to improve recurring donor retention. I chose this model because it was the best performing model after implementing SMOTE.

```python
classifier = RandomForestClassifier(random_state=0) 
classifier.fit(X_train, y_train) 
predictions = classifier.predict(X_test)

feat_importances = pd.Series(classifier.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
```

The graph above shows that the ratio of opportunities to act (i.e. mailings received) to the total number of actions taken (**opps_vs_action_ratio**) was the most influential feature in predicting whether a recurring donor would churn or not. In general, the lower the value, the better, because it would mean people act on majority of the emails they receive from us. This makes logical business sense to be one of the strongest predictors of whether a recurring donor lapses or not, because we would expect more active members who respond to our emails to be less likely to lapse.

Though we can't predict or influence people's actions, we can influence the number of mailings people get and the type of content they get. We can run experiments with varying email volumes as new recurring donors give gifts to see how we can improve recurring donor retention using the optimal email volume. 

## SHAP

We can use the SHAP packages to see how the **random forests model** used and weighted the different features to predict whether an individual recurring donor would lapse or not.

```python
#Randomly selecting a row that might be predicted to be a lapsed recurring donor

# Create object that can calculate shap values
explainer = shap.TreeExplainer(classifier)

# Calculate Shap values
shap_values = explainer.shap_values(X_test.iloc[6])

shap.initjs()
shap.force_plot(explainer.expected_value[1], shap_values[1], X_test.iloc[6])
```

```python
shap.decision_plot(explainer.expected_value[1], shap_values[1], X_test.iloc[6])
```

```python
#Randomly selecting a row that might be predicted to be a not lapsed recurring donor

# Create object that can calculate shap values
explainer = shap.TreeExplainer(classifier)

# Calculate Shap values
shap_values = explainer.shap_values(X_test.iloc[1])

shap.initjs()
shap.force_plot(explainer.expected_value[1], shap_values[1], X_test.iloc[1])
```

```python
shap.decision_plot(explainer.expected_value[1], shap_values[1], X_test.iloc[1])
```

From the SHAP diagrams above, we see that the baseline value of the model is about 0.7, which means the baseline guess for the model is likely to predict any recurring donor as **lapsed**. Additionally, we see how the different features interact together to determine whether the final prediction is lapsed (with more or less confidence), or not lapsed.

For the first person, we see that the **number of mailings in the month before the last gift** was the strongest feature in determining the final prediction for that recurring donor to be lapsed. Though we can't generalize with just one case, it is interesting that because the person got some mailings in the month before their last gift, it made them more likely to lapse. This indicates that email volume is a key feature to observe and also experiment on to see if higher volumes are causing recurring donors to lapse.

In the second case, we see that the gift starting in **2021** was the strongest feature in predicting that the gift would not lapse. Again, though we cannot generalize with one case, it might be likely that the model might be biased to predict that gifts that started in 2021 would not lapse, perhaps because the gift is newer.

Finally, the random forest model highlighted the **opps_vs_action_ratio** as the most important feature earlier on. It is interesting to see that when this metric was lower in the second recurring donor example, it was the second most important feature that influenced the model's **not lapsed** final prediction. When it was higher in the first recurring donor example, it was one of the features that strengthened the model's final **lapsed** prediction for that recurring donor. This is consistent with what was mentioned earlier, which is a good sign.

In summary, from the diagrams above, we can see how the different features interact together for the prediction, and what features were most influential in causing the model to predict the recurring donor as lapsed or not lapsed. 
It is super interesting to see this and how the most influential feature changed, based on the other features! We cannot make any high level generalizations from this alone, but it is a good starting point.
